Our dataset is look like imbalance. first we are going to balance the data than make Prediction on it.

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE()
```

```python
X = email.iloc[:,:-1]
```

```python
y = email.iloc[:,-1]
```

```python
X_final, y_final = smote.fit_resample(X,y)
```

```python
sns.countplot(y_final)
```

as we can see in the above graph, we can clearly say that our dataset is balance now.

```python
email = pd.concat([X_final,y_final], axis = 1)
```

```python
email.shape
```

```python
# Spiliting the data for performing our ml algorithms.
from sklearn.model_selection import train_test_split
```

```python
X = email.iloc[:,:-1]
```

```python
y = email.iloc[:,-1]
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=123)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# Standarize the data
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max = MinMaxScaler()
```

```python
X_train = min_max.fit_transform(X_train)
```

```python
X_test = min_max.transform(X_test)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_pred,y_test)
```

```python
sns.heatmap(confusion_matrix(y_pred, y_test),annot=True, fmt="d", cmap="Blues", linewidths=0.5, square=True)
```

```python
print(classification_report(y_pred, y_test))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rc = RandomForestClassifier()
```

```python
rc.fit(X_train,y_train)
```

```python
y_pred = rc.predict(X_test)
```

```python
accuracy_score(y_pred, y_test)
```

```python
print(confusion_matrix(y_pred,y_test))
```

```python
print(classification_report(y_pred,y_test))
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
accuracy_score(y_pred, y_test)
```

```python
# Choosing K Value
error_rate = []
for i in range(1,30):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    y_pred = knn.predict(X_test)
    err = np.mean(y_pred!=y_test)
    error_rate.append(err)
    
```

```python
error_rate
```

```python
plt.figure(figsize = (15,10))
plt.plot(range(1,30),error_rate,linestyle = 'dashed', marker = '*', markersize = 12, color = 'b', markerfacecolor = 'yellow')
plt.title('K-value Vs Error rate')
plt.xlabel('K-Value')
plt.ylabel('Error Rate')
plt.show()
```

```python
for i in range(1,30):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    print(f'With K = {i}')
    print(f'Confusion Matrix:\n{confusion_matrix(y_pred,y_test)}')
    print()
    print(f'Classification Reports:\n{classification_report(y_pred, y_test)}')
```
